0. *Подготовка базовой модели*

Обучите простую модель классификации с помощью DecisionTreeClassifier на данных из датасета vehicles_dataset_prepared.csv. Для этого сделайте следующие шаги:

1. Обучите модель дерева решений с зафиксированным random_state на тренировочной выборке.

3. Замерьте точность на тестовой выборке и выведите матрицу ошибок.

5. Заново обучите модель и измерьте качество.

````python
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
````

````python
df=pd.read_csv("E:/Sunkar/interesting/Python/data/vehicles_dataset_prepared.csv",sep=",")
x=df.drop(columns=['price_category','price','odometer/price_std'])
y=df['price_category']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
base_dt=DecisionTreeClassifier()
base_dt
````

````python
# Ваш код здесь 
base_dt.fit(x_train, y_train)
preds=base_dt.predict(x_test)
preds_train=base_dt.predict(x_train)
print(accuracy_score(y_test,preds))
print(accuracy_score(y_train,preds_train))
confusion_matrix(y_test,preds)
````

````python
con_df_im=list(zip(x_train.columns,base_dt.feature_importances_))
for i in range(len(con_df_im)):
    if con_df_im[i][1]==0:
        x.drop(con_df_im[i][0],axis=1)
````

````python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
base_dt.fit(x_train, y_train)
preds=base_dt.predict(x_test)
preds_train=base_dt.predict(x_train)
print(accuracy_score(y_test,preds))
print(accuracy_score(y_train,preds_train))
confusion_matrix(y_test,preds)
````
