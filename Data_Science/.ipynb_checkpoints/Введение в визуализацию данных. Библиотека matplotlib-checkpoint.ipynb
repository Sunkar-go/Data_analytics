# Введение в визуализацию данных. Библиотека matplotlib

```python
import pandas as pd
```

```python
df = pd.read_csv('E:/Sunkar/interesting/Python/data/vehicles_dataset_upd.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['odometer']);
```

```python
plt.figure(figsize=(10,5))
plt.hist(df.odometer,bins=2,color='red')
plt.xlabel("Количество пройденных миль")
plt.ylabel("Частота значений")
plt.title("Распределения количества пройденных миль");
```

```python
from matplotlib.ticker import FormatStrFormatter

df_with_emission = pd.read_csv("E:/Sunkar/interesting/Python/data/df_types_upd.csv")
df_with_emission
```

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.hist(df_with_emission.odometer, bins = 100)
ax.set_xlabel("количество пройденных миль")
ax.set_ylabel("Частота значений")
ax.set_title("Распределения количества пройденных миль")
ax.xaxis.set_major_formatter(FormatStrFormatter("%.0f"))
```

```python
fig, (bplt1, bplt2) = plt.subplots(nrows = 2, ncols = 1, figsize = (15,10),sharex=True)
bplt1.hist(df_with_emission.odometer,bins=1000,color='grey')
bplt1.set_title("без аномалий")
bplt1.axvline(df_with_emission.odometer.median(),color="blue",linestyle='dashed',linewidth=1)
bplt2.boxplot(df_with_emission.odometer,vert=False)
bplt2.set_title("с аномалиями")
bplt1.xaxis.set_major_formatter(FormatStrFormatter("%.0f"))
fig.suptitle("Количество пройденных миль");
```
